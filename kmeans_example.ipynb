```python
%matplotlib inline

import matplotlib
import matplotlib.pyplot as plt
import sklearn
import pandas as pd
from sklearn.cluster import KMeans
import numpy as np
```

```python
df = pd.read_csv('sandbox-singlepage/data/housing_prepped.csv',
                 dtype={"GEOID": str,"TRACT_NUM": str})
```

```python
Y = df[['GEOID', 'TOT_POP_2010', 'BORN_FOREIGN_75KPLUS_2010']]
Y = Y[~Y['BORN_FOREIGN_75KPLUS_2010'].isnull()]
```

```python
X = Y[['TOT_POP_2010', 'BORN_FOREIGN_75KPLUS_2010']]
```

```python
K = 2
kmeans = KMeans(n_clusters=K, random_state=0).fit(X)
```

```python
kmeans.labels_
```

```python
kmeans.cluster_centers_
```

```python
Y['labels'] = kmeans.labels_
```

```python
c0 = kmeans.cluster_centers_[0]
c1 = kmeans.cluster_centers_[1]
print(c0, c1)
x0 = np.array([5841.0, 29.0])
print(x0)
np.sqrt(np.sum((x0 - c0)**2)), np.sqrt(np.sum((x0 - c1)**2))
```

```python
c = pd.DataFrame(kmeans.transform(X), columns=['center_{}'.format(i) for i in range(K)])
for i in range(K):
    Y['center_{}'.format(i)] = c['center_{}'.format(i)]
Y
```

```python
for i in range(K):
    Y.loc[Y['labels'] == i, 'd'] = Y['center_{}'.format(i)]
Y
```
