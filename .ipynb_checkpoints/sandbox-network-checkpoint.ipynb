```python
import pandas as pd
import json
import geopandas as gp
from geopy import distance

df= pd.read_csv('sandbox-singlepage/data/housing_prepped.csv',
                   dtype={"GEOID": str,"TRACT_NUM": str})
#edgesdf = pd.read_csv('sandbox-singlepage/data/edges_king_co.csv',  #NO LONGER USING THIS ONE BECAUSE DOES NOT MATCH NUMBER OF EXPECTED PAIRS (~75K for 398 tracts in King Co)
#                   dtype={"GEOID_a": str,"GEOID_b": str})
gdf = pd.read_csv('sandbox-singlepage/data/washingtongeo_dist.csv',
                   dtype={"TRACTCE_a": str,"TRACTCE_b": str})
%matplotlib inline
```

```python
gdf = gdf.merge(df[['TRACT_NUM','GEOID']], how='left', left_on='TRACTCE_a', right_on='TRACT_NUM')
gdf = gdf.rename(columns={'GEOID':'GEOID_a'})
gdf = gdf.merge(df[['TRACT_NUM','GEOID']], how='left', left_on='TRACTCE_b', right_on='TRACT_NUM')
gdf = gdf.rename(columns={'GEOID':'GEOID_b'})
```

```python
gdf['distance'].hist(bins=100)
```

```python
df['BORN_FOREIGN_75KPLUS_PCT_2010'].hist(bins=50)
```

```python
#merge dataframes to combine the different datasets so that you can calculate it.

techie10 = df[['GEOID','BORN_FOREIGN_75KPLUS_PCT_2010']]
gdf = gdf.merge(techie10, how = 'inner', left_on = ['GEOID_a'], right_on = ['GEOID'])
gdf = gdf.rename(columns = {'BORN_FOREIGN_75KPLUS_PCT_2010':'FOREIGN75K_2010_a'})
gdf = gdf.merge(techie10, how = 'inner', left_on = ['GEOID_b'], right_on = ['GEOID'])
gdf = gdf.rename(columns = {'BORN_FOREIGN_75KPLUS_PCT_2010':'FOREIGN75K_2010_b'})

#calculate diff between the two tracts (and take absolute value since sign is meaningless here)
gdf['FOREIGN75K_2010_delta'] = (gdf.FOREIGN75K_2010_a - gdf.FOREIGN75K_2010_b).abs()
```

```python
#delete unnecessary columns to save memory
del gdf['TRACT_NUM_x']
del gdf['TRACT_NUM_y']
del gdf['GEOID_x']
del gdf['GEOID_y']
```

```python
gdf['FOREIGN75K_2010_delta'].hist(bins=100)
```

```python
alpha = .1
gdf['omega'] = (alpha*gdf.FOREIGN75K_2010_delta + (1.0-alpha)*gdf.distance*10e-04)
gdf['omega'].hist(bins=100)
```

```python
ahouse = pd.read_csv('sandbox-singlepage/data/affordable_housing_units.csv',
                   dtype={"TRACT_NUM": str})
```

```python
affhouse = ahouse[(ahouse['CENSUS_QUERY'] == 'B25104_002E')]
affhouse = affhouse.rename(columns = {'DATA' : 'DATA100'})
affhouse200 = ahouse[(ahouse['CENSUS_QUERY'] == 'B25104_003E')]
affhouse200 = affhouse200.rename(columns = {'DATA' : 'DATA200'})
affhouse300 = ahouse[(ahouse['CENSUS_QUERY'] == 'B25104_004E')]
affhouse300 = affhouse300.rename(columns = {'DATA' : 'DATA300'})
affhouse400 = ahouse[(ahouse['CENSUS_QUERY'] == 'B25104_005E')]
affhouse400 = affhouse400.rename(columns = {'DATA' : 'DATA400'})
affhouse500 = ahouse[(ahouse['CENSUS_QUERY'] == 'B25104_006E')]
affhouse500 = affhouse500.rename(columns = {'DATA' : 'DATA500'})
affhouse600 = ahouse[(ahouse['CENSUS_QUERY'] == 'B25104_007E')]
affhouse600 = affhouse600.rename(columns = {'DATA' : 'DATA600'})
affhouse = affhouse.merge(affhouse200, how = 'inner', left_on = ['TRACT_NUM', 'COUNTY','YEAR','DATASET'], right_on = ['TRACT_NUM', 'COUNTY','YEAR','DATASET'])
affhouse = affhouse.merge(affhouse300, how = 'inner', left_on = ['TRACT_NUM', 'COUNTY','YEAR','DATASET'], right_on = ['TRACT_NUM', 'COUNTY','YEAR','DATASET'])
affhouse = affhouse.merge(affhouse400, how = 'inner', left_on = ['TRACT_NUM', 'COUNTY','YEAR','DATASET'], right_on = ['TRACT_NUM', 'COUNTY','YEAR','DATASET'])
affhouse = affhouse.merge(affhouse500, how = 'inner', left_on = ['TRACT_NUM', 'COUNTY','YEAR','DATASET'], right_on = ['TRACT_NUM', 'COUNTY','YEAR','DATASET'])
affhouse = affhouse.merge(affhouse600, how = 'inner', left_on = ['TRACT_NUM', 'COUNTY','YEAR','DATASET'], right_on = ['TRACT_NUM', 'COUNTY','YEAR','DATASET'])
affhouse['AFF_HOUSING_UNITS_2010'] = affhouse['DATA100'] + affhouse['DATA200'] + affhouse['DATA300'] + affhouse['DATA400'] + affhouse['DATA500'] + affhouse['DATA600']
affordable_housing = affhouse[['COUNTY','TRACT_NUM','AFF_HOUSING_UNITS_2010']]
#ahouse.head(2)
```

```python
race = pd.read_csv('sandbox-singlepage/data/race-data.csv',
                   dtype={"TRACT_NUM": str})
```

```python
white = race[(race['CENSUS_QUERY'] == 'B03002_003E')]
white = white[['COUNTY','TRACT_NUM','DATA']]
white = white.rename(columns = {'DATA' : 'pop_white_nonhisp_only'})
black = race[(race['CENSUS_QUERY'] == 'B02001_003E')]
black = black[['COUNTY','TRACT_NUM','DATA']]
black = black.rename(columns = {'DATA' : 'pop_black_only'})
native = race[(race['CENSUS_QUERY'] == 'B02001_004E')]
native = native[['COUNTY','TRACT_NUM','DATA']]
native = native.rename(columns = {'DATA' : 'pop_native_only'})
asian = race[(race['CENSUS_QUERY'] == 'B02001_005E')]
asian = asian[['COUNTY','TRACT_NUM','DATA']]
asian = asian.rename(columns = {'DATA' : 'pop_asian_only'})
polynesian = race[(race['CENSUS_QUERY'] == 'B02001_006E')]
polynesian = polynesian[['COUNTY','TRACT_NUM','DATA']]
polynesian = polynesian.rename(columns = {'DATA' : 'pop_polynesian_only'})
latino = race[(race['CENSUS_QUERY'] == 'B03002_012E')]
latino = latino[['COUNTY','TRACT_NUM','DATA']]
latino = latino.rename(columns = {'DATA' : 'pop_hispanic'})
other = race[(race['CENSUS_QUERY'] == 'B02001_007E')]
other = other[['COUNTY','TRACT_NUM','DATA']]
other = other.rename(columns = {'DATA' : 'pop_other_only'})
multiracial = race[(race['CENSUS_QUERY'] == 'B02001_008E')]
multiracial = multiracial[['COUNTY','TRACT_NUM','DATA']]
multiracial = multiracial.rename(columns = {'DATA' : 'pop_multiracial'})
```

```python
racial = white.merge(black, how = 'inner', left_on = ['TRACT_NUM', 'COUNTY'], right_on = ['TRACT_NUM', 'COUNTY']).drop_duplicates()
racial = racial.merge(native, how = 'inner', left_on = ['TRACT_NUM', 'COUNTY'], right_on = ['TRACT_NUM', 'COUNTY']).drop_duplicates()
racial = racial.merge(asian, how = 'inner', left_on = ['TRACT_NUM', 'COUNTY'], right_on = ['TRACT_NUM', 'COUNTY']).drop_duplicates()
racial = racial.merge(polynesian, how = 'inner', left_on = ['TRACT_NUM', 'COUNTY'], right_on = ['TRACT_NUM', 'COUNTY']).drop_duplicates()
racial = racial.merge(latino, how = 'inner', left_on = ['TRACT_NUM', 'COUNTY'], right_on = ['TRACT_NUM', 'COUNTY']).drop_duplicates()
racial = racial.merge(other, how = 'inner', left_on = ['TRACT_NUM', 'COUNTY'], right_on = ['TRACT_NUM', 'COUNTY']).drop_duplicates()
racial = racial.merge(multiracial, how = 'inner', left_on = ['TRACT_NUM', 'COUNTY'], right_on = ['TRACT_NUM', 'COUNTY']).drop_duplicates()
```

```python
race_data = racial[['COUNTY','TRACT_NUM','pop_white_nonhisp_only','pop_black_only','pop_native_only','pop_asian_only','pop_polynesian_only','pop_hispanic','pop_other_only','pop_multiracial']]
race_data.head(2)
```

```python
df = df.merge(race_data, how = 'inner', left_on = ['TRACT_NUM', 'COUNTY'], right_on = ['TRACT_NUM', 'COUNTY']).drop_duplicates()
df['minority_pop'] = df['TOT_POP_2010'] - df['pop_white_nonhisp_only']
df['minority_pop_pct'] = df['minority_pop'] / df['TOT_POP_2010']
df.head(5)
```
